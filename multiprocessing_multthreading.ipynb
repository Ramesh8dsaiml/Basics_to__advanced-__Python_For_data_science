# what is process 
- A process is a running program with its own independent memory space. means har program apni alag memory space mein chalta hai.


# what is thread:  
- A thread is a lightweight unit of execution inside a process that shares memory.
- Multiple threads within the same process can access shared data and resources.
- means ek process ke anadar kai threads chalte hain jo ek hi memory space share karte hai 

# MULTIPROCESSING (CPU bound):
- Multiprocessing involves using multiple processes to perform tasks concurrently.
- Each process has its own memory space and can run on a separate CPU core, allowing for true parallel execution.
- 

# Multiple processes banana

- har process ki alag memory

- har ek alag CPU core use karta hai

- true parallel execution



- syntax: it usethreading module
- like import multiprocessing

# uses:
- CPU heavy kaam

- large calculations

- ML / DL

- image processing

- data analysis

- One-line yaad rakh:

- Multiprocessing = CPU power ka full use


# Multithreading (I/O bound):
- Multithreading involves using multiple threads within a single process to perform tasks concurrently.
- threads ek hi memory share karte hain
- threads ek hi CPU core use karte hain
- concurrent execution but not true parallelism
- uses: 
  - I/O heavy tasks
  - Web scraping
  - File operations
  - Network requests

  - syntax: it uses threading module
  - like import threading

# GIL = Global Interpreter Lock:
- The Global Interpreter Lock allows only one thread to execute Python bytecode at a time, limiting CPU-bound multithreading.
