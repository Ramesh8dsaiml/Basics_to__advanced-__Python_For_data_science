```python
# ABSTRACTION in Python
# Abstraction means showing only essential features while hiding internal implementation.
# Abstraction reduces complexity and improves usability and security.
# 1. Abstract Class
# 2️. Abstract Method

from abc import ABC, abstractmethod
```

```python
from abc import ABC, abstractmethod

class MLModel(ABC):
    @abstractmethod
    def train(self):
        pass

    @abstractmethod
    def predict(self):
        pass
```

```python
class LinearRegression(MLModel):
    def train(self):
        print("Training Linear Regression")

    def predict(self):
        print("Predicting values")
```

```python
model = LinearRegression()
model.train()
model.predict()
```

```python
class Employee:
    def __init__(self):
        self.__salary = 0   # private
    def set_salary(self, amount):
          if amount > 0:
            self.__salary = amount
          else:
            print("Salary must be positive")
    def get_salary(self):
        return self.__salary
    
```

```python
e = Employee()

e.set_salary(50000)      # valid
print(e.get_salary())

e.set_salary(-1000)      # invalid
```

```python
class Dataset:
    def __init__(self):
        self.__rows = 0

    def set_rows(self, rows):
        if rows > 0:
            self.__rows = rows

    def get_rows(self):
        return self.__rows
```
