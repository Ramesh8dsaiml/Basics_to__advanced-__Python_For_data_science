```python
# dunder function
class complex:
     def __init__(self , real ,imge ):
          self.real=real 
          self.imge = imge
     def shownumber(self):
          print(self.real ,"i+" , self.imge ,"j")

     def __add__(self , num2):                            # use dunder function , use __add__ 
          newreal =self.real  + num2.real 
          newimge =self.imge  + num2.imge
          return complex(newreal , newimge)  

num1 = complex(1,3)
num2 = complex(4,5)
# num3= complex(8,9)
num1.shownumber()
num2.shownumber()
num3 = num1 +num2
num3.shownumber()
```

```python
# dunder function
class complex:
     def __init__(self , real ,imge ):
          self.real=real 
          self.imge = imge
     def shownumber(self):
          print(self.real ,"i+" , self.imge ,"j")

     def __sub__(self , num2):                            # use dunder function , use __add__ 
          newreal =self.real  - num2.real 
          newimge =self.imge  - num2.imge
          return complex(newreal , newimge)  

num1 = complex(1,3)
num2 = complex(4,5)
# num3= complex(8,9)
num1.shownumber()
num2.shownumber()
num3 = num1 -num2
num3.shownumber()
```

```python
class Model:
    def __init__(self, algo, lr):
        self.algo = algo
        self.lr = lr

    def __eq__(self, other):
        return self.algo == other.algo and self.lr == other.lr

m1 = Model("LinearRegression", 0.01)
m2 = Model("LinearRegression", 0.01)

print(m1 == m2)
```

```python
class MLModel:
    def __init__(self, name, lr):
        self.name = name
        self.lr = lr

    def __call__(self, x):
        return x * self.lr

    def __eq__(self, other):
        return self.name == other.name and self.lr == other.lr

    def __str__(self):
        return f"Model: {self.name}, LR: {self.lr}"
```

```python
m1 = MLModel("LinearRegression", 0.1)
m2 = MLModel("LinearRegression", 0.1)

print(m1)
print(m1(10))
print(m1 == m2)
```

```python
# decorators: “Decorator ek function hota hai jo kisi dusre function ke behavior ko bina modify kiye enhance karta hai. Python me decorators possible hain kyunki functions first-class objects hote hain.”
# note: 
# “A decorator is a function that modifies another function’s behavior without changing its source code.”
# user define function decorators :;

def my_decorator(func):
    def wrapper():
        print("Before function")
        func()
        print("After function")
    return wrapper

def hello():
    print("Hello Data Science")

hello = my_decorator(hello)
hello()
```

```python
def my_decorator(func):
    def wrapper():
        print("Before function")
        func()
        print("After function")
    return wrapper

@my_decorator
def hello():
    print("Hello Data Science")

hello()
```

```python
# decorators with arguments:

def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator

@repeat(5)
def train():
    print("Training model")
train()
```

```python
class Student:
    def __init__(self, marks):
        self._marks = marks

    @property
    def marks(self):
        return self._marks

    @marks.setter
    def marks(self, value):
        if value < 50:
            print("Marks cannot be negative")
        else:
            self._marks = value

s = Student(80)
s.marks = 95
print(s.marks)
```
