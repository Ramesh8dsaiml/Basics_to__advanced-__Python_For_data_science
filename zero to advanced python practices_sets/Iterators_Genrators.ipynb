# What is an Iterable?

An iterable is an object that:

Can be looped over

Contains multiple values

Implements the __iter__() method

Examples:

list, tuple, string, set, dictionary

# What is Iteration?

Iteration is the process of:

Accessing elements one by one

Using loops like for or while

# What is an Iterator?

An iterator:

Produces values one at a time

Uses next()

Raises StopIteration when exhausted

# Relationship

Iterable → source of data

Iterator → tool to fetch data step-by-step

Iteration → process of fetching data

# Why Important?

Efficient memory usage

Handling large datasets

Core concept behind generators & pipelines

# Generator Methods

 Generators have special methods for advanced control:

1. send() Method

The send() method allows you to send a value to the generator:

2. close() Method

The close() method stops the generator:

```python
def mygen():
    yield 1
    yield 2
    yield 3
    yield "ramesh"
    yield "ganesh"


```

```python
g = mygen()
print(g)
```

```python
def large_sequence(n):
  for i in range(n):
    yield i

# This doesn't create a million numbers in memory
gen = large_sequence(1000000)
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
def echo_generator():
  while True:
    received = yield
    print("Received:", received)

gen = echo_generator()
next(gen) # Prime the generator
gen.send("Hello")
gen.send("World")
```

```python
def my_gen():
  try:
    yield 1
    yield 2
    yield 3
  finally:
    print("Generator closed")

gen = my_gen()
print(next(gen))
gen.close()
# gen.send(5)
```
