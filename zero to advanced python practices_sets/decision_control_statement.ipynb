Python Data Types :

A data type defines what kind of value a variable holds.

All Data Types Used in Python

Python provides several built-in data types, grouped into categories:

Main Categories:

Numeric → int, float, complex

Boolean → bool

Text → str

Sequence → list, tuple, range

Mapping → dict

Set → set, frozenset

Binary → bytes, bytearray, memoryview

Special → NoneType

 Most Frequently Used Types

int

These cover almost all real-world Python programs.

1: Integer (int)

Integer is a numeric data type which represents whole numbers without decimal.
It can be positive, negative or zero.

a = 10
b = -5

2: Float (float)

Float is a numeric data type used to represent numbers with decimal points.

x = 3.14
y = 2.0


3: Complex (complex)

Complex data type represents numbers in the form a + bj,
where a is real part and b is imaginary part.

c = 2 + 3j

4: Boolean (bool)

Boolean data type represents logical values True or False.

a = True
b = False


5: String (str)

String is an ordered sequence of characters.
It is defined within single quotes (' '), double quotes (" ") or triple quotes.

s = "Python"

6: List (list)

List is an ordered sequence of items same as a tuple,
but it is mutable.
It is defined within square brackets [],
where items are separated by commas.

lst = [1, 2, 3, 4]

7: Tuple (tuple)

Tuple is an ordered sequence of items same as a list,
but it is immutable.
It is defined within parentheses (),
where items are separated by commas.

t = (1, 2, 3)

8: Set (set)

Set is an unordered collection of unique items.
It is defined within curly braces {}.

s = {1, 2, 3}

9: Dictionary (dict)

Dictionary is an unordered collection of key-value pairs.
It is defined within curly braces {},
where each key and value are separated by a colon (:).

d = {"name": "Ramesh", "age": 21}

10: Range (range)

Range represents a sequence of numbers commonly used in loops.

r = range(1, 5)

11: Bytes (bytes)

Bytes is an immutable sequence of integers (0–255).

b = bytes([65, 66, 67])

12: Bytearray (bytearray)

Bytearray is same as bytes but mutable.

ba = bytearray([1, 2, 3])

13: NoneType (None)

None represents the absence of a value.

x = None

Start Decision control statements:

Types of Decision Control Statements in Python

if statement – checks a single condition

if-else statement – handles two possible outcomes

if-elif-else statement – checks multiple conditions

Nested if statement – if statement inside another if

1. if statement:-

if condition:

    statement

working :

Condition True → code run

Condition False → skip

2. if-else Statement

The if-else statement executes one block if condition is true and another block if it is false.

 Syntax:

if condition:

    statement

else:

    statement


3. if-elif-else Statement

Used when multiple conditions need to be checked.

Syntax: 

if condition1:

    statement

elif condition2:

    statement

else:

    statement

4. Nested if Statement


When an if statement is written inside another if statement.

Basic loops:

A loop is used to execute a block of code repeatedly as long as a condition is true.

Types of Loops in Python

1️ while loop

2️ for loop

3 nested loop

1. while loop:

Definition

While loop tab tak code repeat karta hai jab tak condition true hoti hai.

 Syntax

while condition:
    statement

2.  for loop

Definition

For loop kisi sequence (list, string, range) ke elements par iterate karta hai.

 Syntax

for variable in sequence:
    statement

3. nested loop

A nested loop is a loop inside another loop.

Outer loop controls repetitions

Inner loop executes fully for each outer iteration

Key Points

Inner loop completes first

Used in pattern printing, matrices, tables

Can increase time complexity

for i in range(1, 4):
  
    for j in range(1, 3):

        print(i, j)


What is range()?
