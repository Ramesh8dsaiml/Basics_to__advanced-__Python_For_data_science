# Recursive Function :

- recursive function to call itself to solve a problems 

- Two key components :
1. base case : stops recursion
2. recursive case : function call itself 

```python
def count (n):
     if n==0:
          return
     print(n)
     count(n-1)
```

```python
count(5)
```

```python
#factorials 
def factorial(n):
    if n == 1 or n==0 :
        return 1
    return n * factorial(n-1)
```

```python
factorial(5)
```

```python
factorial(4)
```

```python
factorial(6)
```

```python
def fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fib(n-1) + fib(n-2)
```

```python
fib(5)
```

```python
fib(10)
```

```python
# Calculate the sum of all elements in a list:

def sum_list(numbers):
  if len(numbers) == 0:
    return 0
  else:
    return numbers[0] + sum_list(numbers[1:])

my_list = [1, 2, 3, 4, 5]
print(sum_list(my_list))
```

```python

def sum_list(numbers):
  if len(numbers) == 0:
    return 0
  else:
    return numbers[0] + sum_list(numbers[1:3])

my_list = [1, 2, 3, 4, 5]
print(sum_list(my_list))
```

```python
# Find the maximum value in a list:

def find_max(numbers):
  if len(numbers) == 1:
    return numbers[0]
  else:
    max_of_rest = find_max(numbers[1:9])
    return numbers[0] if numbers[0] > max_of_rest else max_of_rest

my_list = [3, 7, 2, 9, 1,54,25,90,87,100,200]
print(find_max(my_list))
```
