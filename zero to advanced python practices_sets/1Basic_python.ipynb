What are membership operators?

Membership operators check whether a value exists in a sequence or collection.

two types membership operatoes 1- in  , 2- not in operatoes 

it is apply on list , tuple , string 


```python
# membership operators

my_list = [1, 2, 3, 4, 5]
print(3 in my_list)      # True
print(6 in my_list)      # False

string = "Hello, World!"
print('H' in string)     # True
```

```python
# List of students
students = ["Ramesh", "Amit", "Suresh", "Neha"]

# Check if student exists
print("Ramesh" in students)      # True ‚Üí Ramesh list me hai
print("Rahul" in students)       # False ‚Üí Rahul list me nahi hai

# -----------------------------------
# Using NOT IN
# -----------------------------------

print("Rahul" not in students)   # True
print("Neha" not in students)    # False

# -----------------------------------
# Membership with string
# -----------------------------------

course = "Python Programming"

print("Python" in course)        # True ‚Üí substring hai
print("Java" not in course)      # True

# -----------------------------------
# Membership with tuple
# -----------------------------------

marks = (45, 67, 89, 90)

print(67 in marks)               # True
print(100 not in marks)          # True

# -----------------------------------
# Membership with set
# -----------------------------------

subjects = {"Math", "Physics", "Chemistry"}

print("Math" in subjects)        # True
print("Biology" not in subjects) # True

```

What are identity operators?

Identity operators check whether two variables refer to the same object in memory.

python support tow types identity operators 

1- is 

2- is not 


```python
a = [1, 2, 3]
b = [1, 2, 3]

print(a == b)   # True
print(a is b)   # False
```

```python
# -------------------------------
# 1Ô∏è Integer identity
# -------------------------------
a = 10
b = 10

# Python small integers ko cache karta hai
print(a is b)        # True ‚Üí same memory
print(a == b)        # True ‚Üí same value


# -------------------------------
# 2Ô∏è Larger integers
# -------------------------------
x = 1000
y = 1000

print(x is y)        # May be False
print(x == y)        # True

# -------------------------------
# 3Ô∏è List identity
# -------------------------------
list1 = [1, 2, 3]
list2 = [1, 2, 3]

print(list1 == list2)    # True ‚Üí same data
print(list1 is list2)    # False ‚Üí different memory

# -------------------------------
# 4Ô∏è Reference copy
# -------------------------------
list3 = list1            # No new list created
print(list1 is list3)    # True

# -------------------------------
# 5Ô∏è Proper copy
# -------------------------------
list4 = list1.copy()
print(list1 == list4)    # True
print(list1 is list4)    # False

# -------------------------------
# 6Ô∏è Identity with None (BEST PRACTICE)
# -------------------------------
value = None

print(value is None)     # Correct way
print(value == None)     # Works but not recommended

# -------------------------------
# 7Ô∏è is not operator
# -------------------------------
print(list1 is not list2)  # True
```
