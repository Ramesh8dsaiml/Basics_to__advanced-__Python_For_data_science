# Lambda Function:

- ontains only one expression 
- using the lambda keywords 
- automatically return the result 
# syntax:
- lambda arguments: expression

- lambda function>> anonymous function
- syntax : lambda arguments : expression >>  
- no need of return statement and def keyword

```python
# Multiply argument a with argument b and return the result:..

x = lambda a, b : a * b
print(x(5, 6))
```

```python
# sub argument a with argument b and return the result:

x = lambda a, b : a - b
print(x(5, 6))
```

```python
# filter
nums = [1, 2, 3, 4, 5, 6]

result = list(filter(lambda x: x % 2 == 0, nums))
print(result)
```

```python
names = ["ram", "shyam", "ramesh", "om"]

result = list(filter(lambda x: len(x) > 4, names))
print(result)
```

```python
# reduce 
from functools import reduce

nums = [1, 2, 3, 4]

result = reduce(lambda a, b: a + b, nums)
print(result)
```

```python
nums = [10, 45, 23, 90]

max_num = reduce(lambda a, b: a if a > b else b, nums)
print(max_num)
```

```python
#sorted 

nums = [-10, 5, -3, 2]

result = sorted(nums, key=lambda x: (x))
print(result)
```

```python
students = [("Ram", 80), ("Shyam", 60), ("Amit", 90)]

result = sorted(students, key=lambda x: x[1])
print(result)
```

```python
#the highest values at each stage you are comparing x and y
numbers = [1, 2, 5, 3, 100, 1000, 6, 10000]
reduce(lambda x, y: x if x>y else y, numbers)
```

```python
def factorial(n):
    return reduce(lambda x, y: x*y, range(1, n+1))
```

```python
factorial(10)
```

```python
factorial(8)
```

```python
factorial(4)
```
